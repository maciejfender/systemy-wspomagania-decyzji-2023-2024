```python
import matplotlib.pyplot as plt
import json
```

```python
dirs = ["Arytmia","Arytmia N","Income","Income N","Iris","Iris N"]
```

```python
files = ["Cartesian.json", "Czebyszew.json","Manhattan.json", "Mahalanobis.json"]
```

```python
dirs_series = {}
dirs_x = {}
for d in dirs:
    dirs_series[d] = {}
    dirs_x[d] = []
    for f in files:
        with open(d+"\\"+f,'r') as file:
            content = json.load(file)
            dirs_series[d][f[:-5]]=list(content.values())
            dirs_x[d] = list(content.keys())
dirs_series
```

```python
dirs_series
```

```python
for series_name, values in dirs_series.items():

    X = list(map(int,dirs_x[series_name]))
    for metric_name,data in values.items():
        plt.title(series_name)
        plt.plot(X,data, label=metric_name)
    plt.xlabel("K parameter")
    plt.ylabel("Correctly Classified")
    plt.legend()
    plt.show()
```
